```python
import pdfplumber
import pandas as pd
```

```python
pdf=pdfplumber.open("moe_report.pdf")
```

```python
tables_array = []
for page in range(len(pdf.pages)):
    p0 = pdf.pages[page]
    #im = p0.to_image()
    tables=p0.extract_tables()
    if tables:
        for sublist in tables:
            if sublist:
                tables_array.append(sublist)
```
